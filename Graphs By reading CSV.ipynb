```python
# Establishments Popular in LA
import matplotlib.pyplot as plt
# pandas for fetching the required data
import pandas as pd

# Read data:
df = pd.read_csv("rest_data_us.csv")

# here, object_type is the List of establishment
# Get the index of object_type and their numbers as x and y (Unique names of all the Establishment names and their numbers)
x = df.groupby("object_type").count().index
y = df.groupby("object_type").count().number
fig = plt.figure()
plt.bar(x,y,color="hotpink")
plt.title("Establishments Popular in LA")
plt.xlabel("Establishments")
plt.ylabel("Numbers")
plt.show()
```

```python
# Proportions of Chain vs Non-chain Esatblishment


labels = []
number = df.groupby("chain").count().number

for i in number.index:
    if i == False:
        labels.append("Establishments without chains")
    else:
        labels.append("Establishments with chains")

plt.title("Proportions of Chain vs Non-chain Esatblishment")
plt.pie(number, labels=labels, autopct= "%1.2f%%")
plt.show()
# display(x)
```
