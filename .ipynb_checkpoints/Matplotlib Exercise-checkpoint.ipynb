```python
import matplotlib.pyplot as plt
import pandas as pd
# Read Total profit of all months and show it using a line plot
data = pd.read_csv("company_sales_data.csv")
display(data)

x = data.month_number
y = data.total_profit
# First plot the x and y values
plt.plot(x,y)
# Set the labels
plt.xlabel("Month Number")
plt.ylabel("Total Profit")
# Set the ticks
plt.xticks(x) # xticks and yticks are the length of values in their corresponding axis which we set.
plt.yticks([100000, 200000, 300000, 400000, 500000])

plt.title("Company profit per month")

plt.show()
```

```python
# Get total profit of all months and show line plot with the following Style properties
x = data.month_number
y = data.total_profit

# Place the legend label in the plot function:

plt.plot(x,y, label = 'Profit data of last year', 
      color='r', marker='o', markerfacecolor='k', 
      linestyle='--', linewidth=3)

plt.xlabel("Month Number")
plt.ylabel("Total Profit")
plt.xticks(x)
plt.yticks([100000, 200000, 300000, 400000, 500000])
# Lets set the legend to identify the markings:

plt.legend(loc="lower right")
plt.show()
```

```python
# Read all product sales data and show it  using a multiline plot

x = data.month_number
facecreamY = data.facecream
facewashY = data.facewash
toothpasteY = data.toothpaste
bathingsoapY = data.bathingsoap
shampooY = data.shampoo
moisturizerY = data.moisturizer


plt.plot(x, facecreamY, marker="o", label="Face Cream Sales Data", linewidth=3)
plt.plot(x, facewashY, marker="o", label="Face Wash Sales Data", linewidth=3)
plt.plot(x, toothpasteY, marker="o", label="Toothpaste Sales Data", linewidth=3)
plt.plot(x, bathingsoapY, marker="o", label="Bathing soap Sales Data", linewidth=3)
plt.plot(x, shampooY, marker="o", label="Shampoo Sales Data", linewidth=3)
plt.plot(x, moisturizerY, marker="o", label="Moisturizer Sales Data", linewidth=3)

plt.xticks(x)
plt.yticks([1000, 2000, 4000, 6000, 8000, 10000, 12000, 15000, 18000])
plt.xlabel("Month Number")
plt.ylabel("Sales units in number")
plt.title("Sales data")
plt.show()
```

```python
x= data.month_number
y = data.toothpaste

plt.scatter(x,y, label="Tooth paste sales data")
plt.legend(loc="upper left")
plt.grid(color="grey", linestyle="--", linewidth=1)
```

```python
x = data.month_number.tolist()
facecreamY = data.facecream
facewashY = data.facewash

plt.bar([z-0.25 for z in x], facecreamY, width=0.25, align="edge", label="Face wash sales data")
plt.bar([z+0.25 for z in x], facewashY, color="orange", width=-0.25, align="edge", label="Face cream sales data")
plt.grid(color="grey", linewidth=0.5, linestyle="--")
plt.xticks(x)
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.legend(loc='upper left')
plt.title(' Sales data')

plt.title('Facewash and facecream sales data')
plt.show()
```
