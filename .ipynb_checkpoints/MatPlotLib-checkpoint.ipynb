```python
from matplotlib import pyplot as plt
import numpy as np
x= np.array([0,10,20,30])
y = np.array([30,40,50,60])
# Creating a line diagram with x and y values
plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel("y-axis")
plt.title("Plotting x and y axis")
plt.show()

# Plotting without values (Drawing big dots):
# We do it with the plot function
plt.plot(x,y, 'o')
plt.show()
# Plotting without values (Drawing dashed lines):
plt.plot(x,y, '--')

# This is how we add multiple markers or fully customize a line it's called string format:
# marker|line|color
plt.show()
```

```python
from matplotlib import pyplot as plt
import numpy as np
import math
x=np.arange(0,math.pi*2,0.05)
print("value of x",x)
y=np.sin(x)
print("value of y",y)
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel("sine")
plt.title("sine wave")
plt.show()
```

```python
import matplotlib.pyplot as plt
y=[1,2,9,16,25,36,49,64]
x1=[1,24,45,67,78,89,99,99]
x2=[34,6,33,45,35,35,34,35]
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
l1=ax.plot(x1,y,'rs-')
l2=ax.plot(x2,y,'go--')
ax.legend(labels=('tv','smartphone'),loc='lower right')
ax.set_title("Advertisement effect on sales")
ax.set_xlabel('medium')
ax.set_ylabel('sales')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import math
x=np.arange(0,math.pi*2,0.05)
fig=plt.figure()
ax=fig.add_axes([0.1,0.1,0.8,0.8])
y=np.sin(x)
ax.plot(x,y)
ax.set_xlabel('angle')
ax.set_title('sine')
ax.set_xticks([0,2,4,6])
ax.set_xticklabels(['zero','two','four','six'])
ax.set_yticks([-1,0,1])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
data = [[30,25,50,20],# Blue vala nu
[40,23,51,17],# Green vala nu
[35,22,45,19]]# Red vala nu
X=np.arange(4)


fig=plt.figure(figsize=(6,6))
ax=fig.add_axes([0,0,1,1])

print(fig)
ax.bar(X+0.00, data[0],color='b',width=0.25)
ax.bar(X+0.25, data[1],color='g',width=0.25)
ax.bar(X+0.50, data[2],color='r',width=0.25)
ax.set_xticks([0.25,1.25,2.25,3.25])
ax.set_xticklabels([2015,2016,2017,2018])
ax.legend(labels=['CS','IT','E&TC'], loc="upper left")
ax.set_title("Mixed bars")
plt.show()
```

```python
fig, ax = plt.subplots(1,1)
bins = [0,25,50,75,100]
a = np.array([22,87,5,43,56,73,55,11,20,51,5,79,31,27])
ax.hist(a, bins= bins)
ax.set_title("Histogram of results")
ax.set_xticks(bins)
ax.set_xlabel("marks")
ax.set_ylabel("No. of students")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

langs = ["C","C++","Java","Python","PHP"]# Labels
students = [23,17,35,29,12]# Values (will be calculated as Percentage) length should match the length of label's array. and the values 
# Should add upto 100
# Value/sum of value * 100


plt.pie(students, labels=langs, autopct= "%1.1f%%")
plt.legend(title="Languages")

plt.show()
```

```python
# Pie diagram for my new priorities:
import matplotlib.pyplot as plt
labels = ["Islam","Family","Future/Work","Love"]
values = [45,25,25,10]
colors = ["lightblue","gold","cyan","red"]
scatter = [0.2, 0.3, 0.1, 0.4]
fig = plt.figure(figsize=(8,8), facecolor="#593", edgecolor="magenta")
ax = fig.add_axes([0,0,1,1])# left, bottom, width, height of the figure

plt.pie(values, labels=labels, colors = colors, shadow=True, explode = scatter, autopct= "%1.1f%%")
ax.set_title("My Priorities")
ax.legend(labels)
# plt.show()

plt.savefig("Priorities.png")
```
